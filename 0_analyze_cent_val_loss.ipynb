```python
import torch.nn as nn
import torch
import utils
import train
import numpy as np
import matplotlib.pyplot as plt
import pickle

device = torch.device("cuda:5")
torch.cuda.is_available(), device
```

```python
BASE_PATH = "/mnt/disk1/mert/FL/saved_models/CIFAR10/CentralizedAccordionResNet56_0/"

def load_data(path):
    with open(path, 'rb') as f:
        data = pickle.load(f)
    return data
    
global_val_losses = load_data(BASE_PATH + 'val_loss.pkl')
global_epoch_losses = load_data(BASE_PATH + 'training_loss.pkl')
global_epoch_accuracies = load_data(BASE_PATH + 'val_acc.pkl')
```

```python
for i in range(len(global_val_losses)):
    global_val_losses[i] = global_val_losses[i].to("cpu").item()
    #global_epoch_accuracies[i] = global_epoch_accuracies[i].to("cpu").item()
```

```python
fig = plt.figure(figsize = (10,5))
plt.plot(np.arange(len(global_val_losses)), global_val_losses, label = "val")
plt.plot(np.arange(len(global_epoch_losses)), global_epoch_losses, label = "train")
plt.legend()
```

```python
plt.plot(np.arange(len(global_epoch_accuracies)), np.array(global_epoch_accuracies), label = "accuracy")
plt.legend()
```
